```python
def next_states(M, r_o):
  r, o = r_o
  with np.errstate(divide = 'ignore', invalid = 'ignore'): d = (o / M)
  m = np.nanmin(d, axis = 1).astype(int)
  m = np.min([m, [1,1,1,1]], axis = 0)
  for a in map(np.array, product(*[list(range(u + 1)) for u in m])):
    op = o - a @ M 
    if op.min() >= 0: 
      yield r + a, r + op
```

```python
from functools import cache

def visit(M, s0):

  maxval = 0

  T = tqdm(leave = False)

  @cache
  def _visit(s, d):
    nonlocal maxval

    #print(d * '\t', s)
    if T is not None: 
      T.set_postfix({'maxval': maxval}, refresh = False)
      T.update()

    if s[1][3] > maxval: maxval = s[1][3]

    if d >= 24: return
    for sp in next_states(M, s):
      _visit(tuple([tuple(sp[0]), tuple(sp[1])]), d + 1)

  _visit(tuple([tuple(s0[0]), tuple(s0[1])]), 0)
```
